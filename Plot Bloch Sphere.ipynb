```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
%matplotlib inline

qc = QuantumCircuit(4)
qc.x(0)
qc.x(2)

state = Statevector.from_instruction(qc)
plot_bloch_multivector(state, title="The state 5 in the computational basis", reverse_bits=False)
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
%matplotlib inline
from numpy import pi

qc = QuantumCircuit(4)
qc.h(0)
qc.h(1)
qc.h(2)
qc.h(3)

qc.p(5*pi/8, 0)
qc.p(5*pi/4, 1)
qc.p(5*pi/2, 2)
qc.p(5*pi/1, 3)

state = Statevector.from_instruction(qc)
plot_bloch_multivector(state, title="The state 5 in the fourier basis", reverse_bits=False)
```
